```python
#4a)
import numpy as np
class Board:     
    won = False
    
    def __init__(self):
        self.board = np.zeros([5,5], dtype = int)
        self.is_marked = np.zeros([5,5], dtype = bool)    
        
    def mark(self, pos_x, pos_y):
        self.is_marked[pos_x, pos_y] = True
    
    def bingo(self):
        if any(np.sum(self.is_marked, 1) == 5):
            return True
        elif any(np.sum(self.is_marked, 0) == 5):
            return True
        else: 
            return False
        
    def score(self, num):
        return num*np.sum((np.ones([5,5], dtype = int) - self.is_marked)*self.board)
```

```python
from collections import deque

def find_parethesis(string):
    
    stack = deque()
    types = [']', ')', '}', ">"]
    for i in range(len(string)):
        if string[i] == '[':
            stack.append(']')
        elif string[i] == '(':
            stack.append(')')
        elif string[i] == '{':
            stack.append('}')
        elif string[i] == '<':
            stack.append('>')
        elif string[i] == stack[-1] == types[0] or string[i] == stack[-1] == types[1] or string[i] == stack[-1] == types[2] or string[i] == stack[-1] == types[3]:
            for p_end in types:
                if string[i] == p_end and stack[-1] == p_end:
                    stack.pop()
        else:
            print("expected", p_end, "but found", string[i], "instead")
            return(string[i])

    return None    
        
```

```python
#10b

def fill_and_score(string):
    score = 0
    stack = deque()
    types = [']', ')', '}', ">"]
    for i in range(len(string)):
        if string[i] == '[':
            stack.append(']')
        elif string[i] == '(':
            stack.append(')')
        elif string[i] == '{':
            stack.append('}')
        elif string[i] == '<':
            stack.append('>')
        elif string[i] == stack[-1] == types[0] or string[i] == stack[-1] == types[1] or string[i] == stack[-1] == types[2] or string[i] == stack[-1] == types[3]:
            for p_end in types:
                if string[i] == p_end and stack[-1] == p_end:
                    stack.pop()
        else:
            #print("expected", p_end, "but found", string[i], "instead")
            return(None)
    
    points = {
        ")": 1,
        "]": 2,
        "}": 3,
        ">": 4,
        "None": 0
    }
    
    for k in range(len(stack)):
        item = stack.pop()
        score *= 5
        score += points[item]
        
    return score   
```

```python
def fold_card(matrix, number, axis):
    if axis == 'y':
        for i in range(np.size(matrix,1)):
            if number - i >= 0 and number + i  <= np.size(matrix,1)-1:
                matrix[:,number - i] += matrix[:, number + i]
                
        matrix = matrix[:, 0:number]    
        
    if axis == 'x':
        for i in range(np.size(matrix,0)):
            if number - i >= 0 and number + i  <= np.size(matrix,0)-1:
                matrix[number - i, :] += matrix[number + i, :]
        
        matrix = matrix[0:number, :]    
        
    return np.array(matrix > 0, dtype = int)
```

```python
#15
with open('C://Users//karol.nycz//Desktop//AdventofCode//input15.txt') as f:
    lines = f.readlines()

lines = [line.replace("\n", "") for line in lines]
lines = [list(line) for line in lines]
lines 
for i in range(len(lines)):
    lines[i] = [int(k) for k in lines[i]]

lines
import numpy as np

graph = np.array(lines)
d = np.zeros([np.size(graph,0), np.size(graph,1)])
p1 = np.zeros([np.size(graph,0), np.size(graph,1)], dtype = int)
p2 = np.zeros([np.size(graph,0), np.size(graph,1)], dtype = int)

def InitSingleSource(graph, start, d):
    for i in range(np.size(graph,0)):
        for j in range(np.size(graph,1)):
            d[i,j] = np.inf
            #p1[i,j], p2[i,j] = 0, 0
    
    d[start[0], start[1]] = 0
    
    return d#, p1, p2
    
def Relax(to_point, from_point, d, graph):
    if d[to_point[0], to_point[1]] > d[from_point[0], from_point[1]] + graph[to_point[0], to_point[1]]:
        d[to_point[0], to_point[1]] = d[from_point[0], from_point[1]] + graph[to_point[0], to_point[1]]
        #p1[to_point[0], to_point[1]] = from_point[0]
        #p2[to_point[0], to_point[1]] = from_point[1]
        
    return d#, p1, p2

def Dijkstra(graph, start, d):
    
    d = InitSingleSource(graph, start, d)
    S = set()
    V = set()
    for i in range(np.size(graph,0)):
        for j in range(np.size(graph,1)):
            V.add((i,j))
    dim_1 = np.size(graph,0)
    dim_2 = np.size(graph,1)
    
    while S != V:
        min_v = None
        
        for node in V - S:
            if min_v != None:
                if d[node] < d[min_v[0], min_v[1]] and node not in S:
                    min_v = node
            else: min_v = node
        
        S.add(min_v)
        if len(S) % 100 == 0: 
            print(len(S), end = " ")
        for i,j in [(1,0), (0,1), (-1,0), (0,-1)]:
            edge = [min_v[0]+i, min_v[1]+j]
            if 0 <= edge[0] < dim_1 and 0 <= edge[1] < dim_2:
                d = Relax(edge, min_v, d, graph)
        
    return d#, p1, p2
```
